## Prática de SQL DDL (esquemas) e DML (instâncias)

Criação e atualização do esquema CARROS2 implantado no SGBD PostgreSQL, seguindo as instruções dos exercícios desta aula 2.

```python
%sql postgresql://postgres:postgres@localhost:5432/postgres
```

```python
# Escolhe o esquema carros como o esquema em que serão feitas as consultas
# (deve-se executar essa célula sempre que for trocar para o esquema carros)
%%sql
SET SCHEMA 'carros';
```

```python
# Exemplo de CREATE TABLE:

# Você pode modificar e adaptar ao teu próprio comando.

%%sql

CREATE TABLE carros.automoveis (
    codigo integer NOT NULL,
    ano character(2) NOT NULL,
    fabricante character(20),
    modelo character(20),
    preco_tabela numeric(8,2),
    pais character(20)
);
```

```python
%%sql

CREATE TABLE carros.revendedoras (
  cnpj char(14) not null,
  nome varchar(45) not null,
  cpf_proprietario char(11) not null,
  estado char(2),
  constraint PK_revendedoras PRIMARY KEY (cnpj)
)
```

```python
%%sql

CREATE TABLE carros.consumidores (
  cpf char(11) not null,
  nome varchar(25) not null,
  sobrenome varchar(25),
  data_nascimento date,
  estado char(2),
  constraint PK_consumidores PRIMARY KEY (cpf)
)
```

```python
%%sql

CREATE TABLE carros.negocios (
  cpf_comprador char(11) not null,
  cnpj_revendedora char(14) not null,
  codigo_automovel integer not null,
  ano character(2) not null,
  data date,
  preco numeric(8,2)
)
```

```python
%%sql

CREATE TABLE carros.garagens (
  cnpj_revendedora char(14) not null,
  codigo_automovel integer not null,
  ano character(2) not null,
  quantidade integer not null
)
```

Alterar as restrições das tabelas de relacionamentos:
- Adicionar chaves primárias;
- Adicionar chaves estrangeiras;

```python
%%sql

ALTER TABLE carros.automoveis
  ADD CONSTRAINT PK_automoveis PRIMARY KEY (codigo, ano);
```

```python
%%sql

ALTER TABLE carros.negocios
  ADD CONSTRAINT PK_negocios PRIMARY KEY (cpf_comprador, cnpj_revendedora, codigo_automovel, ano),
  ADD CONSTRAINT FK_negocios_consumidores FOREIGN KEY (cpf_comprador) REFERENCES Consumidores(cpf),
  ADD CONSTRAINT FK_negocios_revendedoras FOREIGN KEY (cnpj_revendedora) REFERENCES Revendedoras(cnpj),
  ADD CONSTRAINT FK_negocios_automoveis FOREIGN KEY (codigo_automovel, ano) REFERENCES Automoveis(codigo, ano)
```

```python
%%sql

ALTER TABLE carros.garagens
  ADD CONSTRAINT PK_garagens PRIMARY KEY (cnpj_revendedora, codigo_automovel, ano),
  ADD CONSTRAINT FK_garagens_revendedoras FOREIGN KEY (cnpj_revendedora) REFERENCES Revendedoras(cnpj),
  ADD CONSTRAINT FK_garagens_automoveis FOREIGN KEY (codigo_automovel, ano) REFERENCES Automoveis(codigo, ano)
```

```python
%%sql

ALTER TABLE carros.revendedoras
  ADD CONSTRAINT FK_revendedoras_consumidores FOREIGN KEY (cpf_proprietario) REFERENCES Consumidores(cpf)
```
